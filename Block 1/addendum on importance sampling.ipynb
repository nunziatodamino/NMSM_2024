```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Step 1: Define the polynomial function of the form a + bx^2
def poly_func(x, a, b):
    return a + b * x**2

# Step 2: Generate sample data for sin(x)
x_data = np.linspace(0, np.pi/2, 100)  # x values from -pi to pi
y_data = np.sin(x_data)  # Corresponding sin(x) values

# Step 3: Fit the polynomial to the sin(x) data using curve_fit
params, _ = curve_fit(poly_func, x_data, y_data)

# Step 4: Extract the fitted coefficients a and b
a_fit, b_fit = params

# Step 5: Generate the fitted polynomial values
y_fit = poly_func(x_data, a_fit, b_fit)

# Step 6: Plot the original sin(x) function and the fitted polynomial
plt.plot(x_data, y_data, label="sin(x)", color="blue")
plt.plot(x_data, y_fit, label=f"Fitted Polynomial: {a_fit:.3f} + {b_fit:.3f}x^2", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Fitting sin(x) with a Polynomial of the form a + bx^2")
plt.show()
```
