# Sampling point on a D-dimensional volume -  evaluation of  the volume of an ellipsoid

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Ellipsoid parameters
a = 3.0
b = 2.0
c = 2.0

iterationNumber = 10000

# add machine precision check on iterations
```

```python
def analyticEllipsoidOctant (a,b,c):
    return 1/8 *4/3 * np.pi *a * b * c 

def boxVolume (a,b,c):
    return a * b * c

def monteCarloEllipsoidOctant (a,b,c):
    iterations = []
    error = []
    nCounter =0
    for i in range(iterationNumber):
        x = a*random.random()
        y = b*random.random()
        z = c*random.random()
        if (x**2/a**2)+(y**2/b**2)+(z**2/c**2)<1 : nCounter+=1
        if i % 100 == 0 : 
            iterations.append(i)
            volumeEstimate = nCounter/iterationNumber * boxVolume(a,b,c)
            error.append(abs(analyticEllipsoidOctant(a,b,c)-volumeEstimate))
    return iterations, error
```

```python
iterations, error = monteCarloEllipsoidOctant(a,b,c)

plt.plot(iterations, error, marker='o', linestyle='-', color='b')

plt.title('Error of the Monte Carlo simulation in function of the number of iterations')
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.show()
```

### What if we change parameters ?


```python
# New ellipsoid parameters
a1 = 3.0
b1 = 1.0
c1 = 1.0

iterations1, error1 = monteCarloEllipsoidOctant(a1,b1,c1)

plt.plot(iterations, error, marker='o', linestyle='-', color='b')
plt.plot(iterations1, error1, marker='o', linestyle='-', color='r')

plt.title('Error of the Monte Carlo simulation in function of the number of iterations')
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.show()
```
