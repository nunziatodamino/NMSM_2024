We have as pdf:

$$ \rho(x) = A_f \exp\left( -8 \left( \frac{x^2}{2}+\frac{x^4}{4} \right) \right) $$

With the following normalization condition:

$$ \int_{\mathbb{R}} \rho(x) dx = 1 \qquad \iff \qquad   \frac{eA_f}{\sqrt{2}}K_{1/4}(1) = 1$$

Where we denote with $K_{1/4}(x)$ the Bessel function of the second kind.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.special import kv
```

```python
A = np.sqrt(2)/(np.e * kv(1/4,1))
print(A)
```

So we find that $A_f \simeq 1.21$.
Now we want to find an invertible pdf $g(x)$, with the property:

$$ \rho(x) \leq c g(x) $$

The closer the function $cg(x)$ is to $\rho(x)$ the more efficient the sampling is.
We can choose as a pdf $N(0,1/2)$ with c = 1.6, as shown in the picture.

```python
def nonInvertiblePdf (x):
    return A * np.exp ( -8*(x**2/2+x**4/4) )
```

```python
x = np.arange(-3 ,+3, 0.01)
y = nonInvertiblePdf(x)

mean = 0
sigma = np.sqrt(1/4)

c = 1.6

plt.plot (x,y)
plt.plot(x, c *stats.norm.pdf(x, mean, sigma))
plt.show()
```

```python
numberSampledPoints = 100000
goodRandomVariables = []

for i in range (numberSampledPoints):
    randomVariable = np.random.normal(mean,sigma)
    uniformRandomNum = random.random()
    if uniformRandomNum < nonInvertiblePdf(randomVariable)/(c* stats.norm.pdf (randomVariable,mean, sigma)):
        goodRandomVariables.append(randomVariable)
```

```python
plt.plot (x,y)
plt.hist(goodRandomVariables, bins=30, density=True, alpha=0.4, color='r')
plt.show()
```
